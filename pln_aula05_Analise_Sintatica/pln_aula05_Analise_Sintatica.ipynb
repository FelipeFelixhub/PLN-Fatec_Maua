## Inicio do processamento do corpus

```python
# criar uma funcao para fazer a abertura e leitura do aquivo
def ler(nome_arquivo):
  arquivo = open(nome_arquivo, 'r', encoding='utf-8')
  conteudo_arq = arquivo.read()
  arquivo.close()
  return conteudo_arq
```

```python
texto = ler('/content/drive/MyDrive/Colab Notebooks/PNL/2025.05 Ubirajara.txt')
print(len(texto))
```

## Buscador de palavras

```python
def buscador(alvo, texto):
  texto = texto.replace('\n', ' ')
  texto = texto.replace('\t', ' ')

  ocorrencias = []
  encontrado_aqui = texto.find(alvo, 0)

  while encontrado_aqui > 0:
    pos_inicial = encontrado_aqui - (40 - len(alvo))
    trecho = texto[pos_inicial: pos_inicial + 80]

    ocorrencias.append(trecho)

    encontrado_aqui = texto.find(alvo, encontrado_aqui + 1)

  return ocorrencias
```

```python
resultados = buscador('peito', texto)
for resultado in resultados:
  print(resultado)
```

## Segunda Etapa - Limpeza do Corpus

```python
palavras = texto.split()
```

```python
teste = "Corre-se atrás do carro, corre se o carro for embora."
word = teste.split()
print(word)
```

```python
def limpar(lista):
  lixo = '.,:;?!""`()[]{}\|/@#$%^&*~-_'
  quase_limpo = [x.strip(lixo).lower() for x in lista]
  return [x for x in quase_limpo if x.isalpha() or '-' not in x]
```

```python
print(limpar(word))
```

```python
print(len(palavras))
palavras = limpar(palavras)
print(len(palavras))
```

## Exemplo 1 - POS Tagging com Spacy

```python
# Instalação do Spacy
!pip install spacy
!python -m spacy download pt_core_news_sm
```

```python
import spacy

nlp = spacy. load("pt_core_news_sm")

frase = input("Insira uma frase, gramaticalmente correta: ")

doc = nlp(frase)

for token in doc:
  print(f"{token.text} -> {token.pos_}")
```

## Exemplo 2 - POS Tagging com NLTK (somente em ingles)

```python
!pip install nltk
nltk.download("averaged_perceptron_tagger")
nltk.download('punkt_tab')
```

```python
import nltk

frase = input("Insira uma frase, coerente: ")
tokens = nltk.word_tokenize(frase)

pos_tags = nltk. pos_tag(tokens)

for palavra, classe in pos_tags:
  print(f'{palavra}->{classe}')
```

## Exemplo 3 - Parsin de Dependencia com spacy


```python
import spacy

nlp = spacy.load('pt_core_news_sm')

frase = ("O gato preto dorme na cadeira ")

doc = nlp(frase)

for token in doc:
  print(f"{token.text} -> {token.dep_} -> {token.head.text}")
```

## Exemplo 4 - Analise de Dependencia com spacy


```python
import spacy
from spacy import displacy

nlp = spacy.load('pt_core_news_sm' )

frase = "0 gato preto dorme na cadeira."

doc = nlp(frase)

displacy.render(doc, style='dep', jupyter=True)
```

```python
import nltk
from nltk import Tree

sintaxe = Tree('S', [
  Tree('NP', [Tree('DET', ['O']), Tree('N', ['gato'])]),
  Tree('VP', [Tree('V', ['dorme']),
              Tree('PP', [Tree('P',['na']), Tree('NP',[Tree('N', ['cadeira'])])])])
])

sintaxe.pretty_print()
```
